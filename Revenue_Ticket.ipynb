```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the Excel file for the data analysis
file_path = '/content/REVENUE CHECK DEC 2023-JUNE 2024 (1).xlsx'
xls = pd.ExcelFile(file_path)

# Extract sheet names to determine which sheets to load
sheet_names = xls.sheet_names

# Load the Monthly Comparative Analysis and Summary sheets for analysis
monthly_comparative = pd.read_excel(xls, sheet_name='MONTHLY COMPARATIVE ANALYSIS')
summary = pd.read_excel(xls, sheet_name='SUMMARY')

# Display the first few rows of the monthly comparative analysis and summary sheets
monthly_comparative_head = monthly_comparative.head()
summary_head = summary.head()

# Calculate the percentage increase for BOLT, CAR HIRE, and TASK FORCE
initial_bolt = monthly_comparative['BOLT'].iloc[0]
final_bolt = monthly_comparative['BOLT'].iloc[-1]
bolt_increase_percentage = ((final_bolt - initial_bolt) / initial_bolt) * 100

initial_car_hire = monthly_comparative['CAR HIRE'].iloc[0]
final_car_hire = monthly_comparative['CAR HIRE'].iloc[-1]
car_hire_increase_percentage = ((final_car_hire - initial_car_hire) / initial_car_hire) * 100

initial_task_force = monthly_comparative['TASK FORCE'].iloc[0]
final_task_force = monthly_comparative['TASK FORCE'].iloc[-1]
task_force_increase_percentage = ((final_task_force - initial_task_force) / initial_task_force) * 100

# Display the percentage increase
print(f"BOLT Increase Percentage: {bolt_increase_percentage:.2f}%")
print(f"CAR HIRE Increase Percentage: {car_hire_increase_percentage:.2f}%")
print(f"TASK FORCE Increase Percentage: {task_force_increase_percentage:.2f}%")

# Plotting the Monthly Revenue for BOLT, CAR HIRE, and TASK FORCE
plt.figure(figsize=(10, 6))

# Plotting each category
plt.plot(monthly_comparative['MONTHS'], monthly_comparative['BOLT'], label='BOLT', marker='o')
plt.plot(monthly_comparative['MONTHS'], monthly_comparative['CAR HIRE'], label='CAR HIRE', marker='o')
plt.plot(monthly_comparative['MONTHS'], monthly_comparative['TASK FORCE'], label='TASK FORCE', marker='o')

# Adding labels and title
plt.title('Monthly Revenue Analysis (BOLT, CAR HIRE, and TASK FORCE)', fontsize=14)
plt.xlabel('Months', fontsize=12)
plt.ylabel('Revenue (NGN)', fontsize=12)
plt.legend()
plt.grid(True)
plt.tight_layout()

# Save the plot as an image to include in the document
plt_path = "monthly_revenue_analysis.png"
plt.savefig(plt_path)

plt.show()
```
